```python
import numpy as np
import pandas as pd

import seaborn 

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import os
logpath = './build/log1.txt'

f = open (logpath,'r')

line = f.readline()

cte = []
steer = []
totalerror = []
throttle =  []



while line:
    
    log = line;
    
    cte.append(log.split(' ')[1])
    steer.append(log.split(' ')[3])
    totalerror.append(log.split(' ')[5])
    throttle.append(log.split(' ')[7])
    
    line = f.readline()
    
f.close()
```

```python
plt.figure(figsize=(20,6))
plt.plot(cte)
plt.plot([0 for i in range(len(cte))],'r')
#plt.plot(throttle,'g')
#plt.plot([0.5 for i in range(len(throttle))],'r-')
plt.title("Cross Track Error")
plt.xlabel("Steps")
plt.ylabel("Cte")
plt.show()
```

```python
plt.figure(figsize=(20,6))
plt.plot(steer)
plt.plot([0 for i in range(len(steer))],'r')
plt.title("Steer Value")
plt.xlabel("Steps")
plt.ylabel("Steer value")
plt.show()
```

```python
plt.figure(figsize=(20,6))
plt.plot(totalerror)
plt.plot([0 for i in range(len(totalerror))],'r')
plt.title("Total Error")
plt.xlabel("Steps")
plt.ylabel("TotalError")
plt.show()
```

```python
plt.figure(figsize=(20,6))
plt.plot(throttle)
plt.plot([0.5 for i in range(len(throttle))],'r')
plt.title("Throttle Value")
plt.xlabel("Steps")
plt.ylabel("throttle value")
plt.show()
```
